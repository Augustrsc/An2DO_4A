# Présentation synthétique des résultats


Analyses et plots à retenir principalement pour l'oral.

## 1. Nettoyage du jeu de données

Etape préliminaire mais qui a pris du temps : jeu de données de 14 252 établissements proposés sur Parcoursup. 118 variables, dont une grande majorité peu utiles en pratique pour l'analyse. Beaucoup de NaN, on a donc supprimé les variables qui en contenaient trop et mis les autres à la médiane. On a enevé pas mal de variables jusqu'à obtenir un dataset de 31 variables, dont 7 qualitatives et 24 quantitatives.

## 2. Visualisation des données

### 2.1. Analyse uni-dimensionnelle

Histogramme de toutes les var quantitatives, et tableau des fréquences (ou des effectifs) de toutes les var quali. Observations principales (voir le fichier analyses/Analyse_Uni.ipynb): 
- Etablissements divisés en types de formation : BTS, BUT, Licence, LAS, PASS, Ecole d'ingénieur, Ecole de commerce, CPGE, IFSI, EFTS. BTS et Licence majoritaires. Remarque : catégorie "Autres formations" qui regroupe beaucoup d'individus, on sait pas trop ce qu'il y a derrière (il faudrait regarder quelques individus pour voir un peu).
- Filières de formation principalement concentrées dans les régions qui comportent des grandes villes étudiantes (Paris, Lyon, Lille, Toulouse, Bordeaux...), diagramme barre à mettre en relation avec la carte
- Faire la même analyse pour satut public/privé et sélectif/non-sélectif

Variables quantitatives : 
- pourcentages de filles/garçons admis : beaucoup de formations uniquement masculines ou féminines
- taux de remplissage : certaines filières sous-remplies, et d'autres sur-remplies (voir analyse bi-dim pour voir lesquelles)
- pourcentage d'admis issus de la même académie : presque 70% en moyenne (toutes formations confondues) fort décalage vers la droite, ie beaucoup d'établissements ont une grande partie de leurs étudiants qui sont de la même région.
- En moyenne 22.27% d'admis sont des boursiers, toutes formations confondues
